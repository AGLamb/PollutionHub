```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import math
```

```python
Locations = pd.read_csv("../DTO/LocationGrid.csv", header=None).to_numpy()
print(Locations)
```

```python
Tensor = np.load("../DTO/SpatialTensor.npy")
```

```python
def plot_heatmap_with_labels(matrix: np.ndarray, labels: np.ndarray) -> None:
    if matrix.shape[0] != labels.shape[0]:
        raise ValueError("Labels provided do not match matrix dimensions.")

    # Get the maximum values of coordinates in each dimension to define heatmap size
    max_dims = np.max(labels, axis=0) + 1

    # Create an empty array for the heatmap
    heatmap = np.zeros(max_dims, dtype=matrix.dtype)

    # Fill in the values from the first column of the matrix
    for i, coord in enumerate(labels):
        heatmap[tuple(coord)] = matrix[i][0]

    plt.figure(figsize=(10,10))

    # Use seaborn's heatmap function to plot heatmap
    ax = sns.heatmap(heatmap, annot=True, fmt="d")

    plt.show()
    return None
```

```python
plot_heatmap_with_labels(Tensor[20, :, :], Locations)
```
